```python
import os
import pandas
class EmployeesManagement:
  def __init__(self):
    self.employees=[]
    self.load_data()

  def save_csv(self):
    pd.DataFrame(self.employees).to_csv("data.csv")

  def load_data(self):
    if os.path.exists("data.csv"):
      df=pd.read_csv("data.csv")
      for k in range(len(df)):
        self.add_e(df.iloc[k]['name'],df.iloc[k]['salary'],df.iloc[k]['age'],df.iloc[k]['position'])
    else:
      pass

  def add_e(self,name,salary,age,position):
    id=len(self.employees)+1
    self.employees.append({'name':name,'salary':salary,'age':age,'position':position,'id':id})
    print("New Employeees Added Successfully")
  def remove_e_by_id(self,employee_id):
    for emp in self.employees:
      if emp['id']==employee_id:
        self.employees.remove(emp)
        print(f"Employee with ID {employee_id} is removed successfully")
      else:
        print("Employee Id Doesn't Exist in This List")
  def remove_e_by_name(self,employee_name):
    for emp in self.employees:
      if emp['name']==employee_name:
        self.employees.remove(emp)
        print(f"Employee with Name {employee_name} is removed successfully")
      else:
        print("Employee Name Doesn't Exist in This List")
  def update_e_by_id(self,employee_id,name=None,salary=None,age=None,position=None):
    for emp in self.employees:
      if emp['id']==employee_id:
        if name:
          emp['name']=name
        if salary:
          emp['salary']=salary
        if age:
          emp['age']=age
        if position:
          emp['position']=position
        print(f"Employee with ID {employee_id} is Updated Successfully")
  def update_e_by_name(self,employee_name,new_name=None,salary=None,age=None,position=None):
    for emp in self.employees:
      if emp['name']==employee_name:
        if new_name:
          emp['new_name']=new_name
        if salary:
          emp['salary']=salary
        if age:
          emp['age']=age
        if position:
          emp['position']=position
        print(f"Employee with Name {employee_name} is Updated Successfully")
  def display_e(self):
    if self.employees:
      for emp in self.employees:
        print(f"Name:{emp['name']}, Salary:{emp['salary']}, Age:{emp['age']}, Position:{emp['age']}, Id:{emp['id']}")
    else:
      print("No Elements to Display")

e1=EmployeesManagement()
while True:
  print("\n Employees Management System Menu")
  print("1. Add Employees")
  print("2. Remove Employees By Id")
  print("3. Remove Employees By Name")
  print("4. Update Employee By Id")
  print("5. Update Employee By Name")
  print("6. Display All Employees")
  print("7. Exit")
  choice=int(input("Enter the Choice:"))
  if choice==1:
    name=input("Enter the Name of Employee:")
    salary=float(input("Enter the Salary of Employee:"))
    age=int(input("Enter the Age of Employee:"))
    position=input("Enter the Position of Employee:")
    e1.add_e(name,salary,age,position)
  elif choice==2:
    employee_id=int(input("Enter the Id of Employee to Be Removed:"))
    e1.remove_e_by_id(employee_id)
  elif choice==3:
    employee_name=input("Enter the Name of Employee to Be Removed:")
    e1.remove_e_by_name(employee_name)
  elif choice==4:
    employee_id=int(input("Enter the Employee Id To Be Updated:"))
    name=input("Enter the New Name(Press Enter To Skip) :")
    salary=float(input("Enter the New Salary(Press Enter To Skip):"))
    age=input("Enter the New Age(Press Enter To Skip):")
    age=int(age) if age else None
    position=input("Enter the New Position(Press Enter To Skip):")
    e1.update_e_by_id(employee_id)
  elif choice==5:
    employee_name=input("Enter the Current Name to be Updated:")
    new_name=input("Enter the New Name(Press Enter To Skip) :")
    salary=float(input("Enter the New Salary(Press Enter To Skip) :"))
    age=input("Enter the New Age(Press Enter To Skip):")
    age=int(age) if age else None
    position=input("Enter the New Position(Press Enter To Skip):")
    e1.update_e_by_name(employee_name)
  elif choice==6:
    e1.display_e()
  elif choice==7:
    e1.save_csv()
    print("Exit the Program")
    break
  else:
    print("Wrong Choice,Enter the Valid Choice")
```
