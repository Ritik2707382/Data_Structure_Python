```python
import pandas as pd
class EmployeesManagement:
  def __init__(self):
    self.employees=pd.DataFrame(columns=['Name','Salary','Age','Position','ID'])
  def add_e(self,name,salary,age,position):
    employee_id=len(self.employees)+1
    self.employees=self.employees.append({'Name':name ,'Salary':salary,'Age':age,'Position':position,'ID':employee_id},ignore_index=True)
    print(f"New Employees are added successfully")
  def remove_e_by_id(self,employee_id):
    if not self.employees[self.employees['ID']==employee_id].empty:
      self.employees= self.employees[self.employees['ID']!=employee_id]
      print(f"Employee with Id {employee_id} is removed successfully")
    else:
      print("Employee with Id doesn't exist")
  def remove_e_by_name(self,employee_name):
    if not self.employees[self.employees['Name']==employee_name].empty:
      self.employees=self.employees[self.employees['Name']!=employee_name]
      print(f"Employee with Name {employee_name} is removed successfully")
    else:
        print("Employee with Name doesn't exist")
  def update_by_id(self,employee_id,name=None,age=None,position=None,salary=None):
    if not self.employees[self.employees['ID']==employee_id].empty:
      if name:
        self.employees.loc[self.employees['ID']==employee_id, 'Name']=name
      if age:
        self.employees.loc[self.employees['ID']==employee_id, 'Age']=age
      if position:
        self.employees.loc[self.employees['ID']==employee_id, 'Position']=position
      if salary:
        self.employees.loc[self.employees['ID']==employee_id, 'Salary']=salary
    print(f"Employee with Id {employee_id} is updated successfully")
  def update_by_name(self,employee_name,new_name=None,salary=None,age=None,position=None):
    if not self.employees[self.employees['Name']==employee_name].empty:
      if new_name:
        self.employees.loc[self.employees['Name']==employee_name, 'Name']=new_name
      if age:
        self.employees.loc[self.employees['Name']==employee_name, 'Age']=age
      if position:
        self.employees.loc[self.employees['Name']==employee_name, 'Position']=position
      if salary:
        self.employees.loc[self.employees['Name']==employee_name, 'Salary']=salary
    print(f"Employee with Name {employee_name} is updated successfully")
  def display_employee(self):
    if not self.employees.empty:
      print(self.employees)
    else:
      print("No Employees to Display")

e1=EmployeesManagement()
while True:
  print("\n Employee Management System Menu")
  print("1. Add Employee")
  print("2. Remove Employee By ID")
  print("3. Remove Employee By Name")
  print("4. Update Employee By ID")
  print("5. Update Employee By Name")
  print("6. Display All Employees")
  print("7. Exit")
  choice=int(input("Enter the Choice:"))
  if choice==1:
    name=input("Enter the Name:")
    salary=float(input("Enter the Salary:"))
    age=int(input("Enter the Age:"))
    position=input("Enter the Position:")
    e1.add_e(name,salary,age,position)
  elif choice==2:
    id=int(input("Enter the ID to be Removed:"))
    e1.remove_e_by_id(id)
  elif choice==3:
    name=input("Enter the Name to be Removed:")
    e1.remove_e_by_name(name)
  elif choice==4:
    id=int(input("Enter the Id to be Updated:"))
    name=input("Enter the New Name (Press Enter to Skip):")
    salary=input("Enter the New Salary (Press Enter to Skip) :")
    salary=float(salary) if salary else None
    age=input("Enter the New Age(Press Enter to Skip):")
    age=int(age) if age else None
    position=input("Enter the New Position(Press Enter to Skip):")
    e1.update_by_id(id,name,salary,age,position)
  elif choice==5:
    old_name=input("Enter the Current Name of Employee to be Updated:")
    new_name=input("Enter the New Name(Press Enter to Skip):")
    salary=input("Enter the New Salary (Press Enter to Skip):")
    salary=float(salary) if salary else None
    age=input("Enter the New Age(Press Enter to Skip):")
    age=int(age) if age else None
    position=input("Enter the New Position(Press Enter to Skip):")
    e1.update_by_name(old_name,new_name,salary,age,position)
  elif choice==6:
    e1.display_employee()
  elif choice==7:
    print("Exit the Program")
    break
  else:
    print("Wrong Choice- Enter the Valid Choice")
```
